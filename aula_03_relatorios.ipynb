## Relatório de Análise - Hábitos e Desempenho dos Alunos

**"1. Engenharia de Prompt"**
Constatei que a melhor resposta é a 3ª, pois ela é mais clara e objetiva.

Após analisar os dados dos 1.000 alunos, os três hábitos com maior associação à nota final (exam_score) são os seguintes:

Posição	Hábito	Evidência (correlação com a nota)	Interpretação
1	Horas de estudo por dia	0,825	Impacto muito forte e positivo. Quanto mais o aluno estuda, maior tende a ser sua nota.
2	Saúde mental	0,322	Impacto moderado e positivo. Alunos com melhor avaliação de saúde mental apresentam, em média, melhor desempenho.
3	Tempo assistindo Netflix	-0,172	Impacto moderado e negativo. Mais tempo assistindo Netflix está associado a notas mais baixas.
Resumo para a direção da escola

A análise indica que o fator mais importante para o desempenho acadêmico é o tempo dedicado aos estudos, com uma correlação de 0,825, muito superior à dos demais hábitos. Em segundo lugar, a saúde mental apresenta uma associação positiva (0,322), sugerindo que ações de apoio ao bem-estar dos estudantes podem contribuir para melhores resultados. Por fim, o tempo gasto assistindo Netflix possui uma associação negativa (-0,172), indicando que um maior tempo dedicado a essa atividade tende a estar relacionado a um desempenho inferior.

Conclusão: os dados sugerem que iniciativas para estimular uma rotina consistente de estudos, promover o bem-estar emocional dos alunos e incentivar o equilíbrio no uso do tempo livre têm maior potencial para melhorar o desempenho acadêmico nesta base de estudantes.
