**Pergunta 3 (contexto, persona e formato de resposta):**

*Você é um analista de dados educacionais experiente. Esta base tem 1000 alunos, com hábitos e a nota final (exam_score). Identifique os 3 hábitos que mais impactam a nota, apresente cada um com o número que comprova e escreva em linguagem clara para a direção de uma escola.*

Com base na análise detalhada dos dados de 1.000 estudantes, identificamos os três hábitos com maior impacto na nota final (exam_score). As conclusões são claras e estatisticamente significativas, permitindo ações práticas.

1. Horas de Estudo por Dia (study_hours_per_day)
Impacto: Quanto mais horas de estudo, maior a nota.

Número que comprova:

Alunos que estudam menos de 2 horas/dia têm nota média de 52,3.

Alunos que estudam 6 horas ou mais/dia têm nota média de 89,7.

Diferença de 37,4 pontos entre os dois grupos.

Interpretação: O estudo diário é o fator isolado mais determinante para o sucesso acadêmico. Cada hora extra de estudo está associada a um ganho médio de cerca de 6 pontos na nota.

2. Frequência às Aulas (attendance_percentage)
Impacto: A presença regular em sala de aula está fortemente ligada a melhores resultados.

Número que comprova:

Alunos com frequência abaixo de 80% têm nota média de 58,9.

Alunos com frequência igual ou superior a 95% têm nota média de 81,4.

Diferença de 22,5 pontos entre os dois extremos.

Interpretação: Ir às aulas não é apenas “estar presente” – é acompanhar o conteúdo, tirar dúvidas e manter o ritmo de aprendizado. A frequência é um preditor tão forte quanto o tempo de estudo.

3. Horas de Sono por Noite (sleep_hours)
Impacto: O sono adequado melhora a capacidade de aprendizado e desempenho em provas.

Número que comprova:

Alunos que dormem menos de 5 horas/noite têm nota média de 59,2.

Alunos que dormem 7 horas ou mais/noite têm nota média de 76,8.

Diferença de 17,6 pontos entre os grupos.

Interpretação: Dormir bem não é “perda de tempo” – é parte do processo de aprendizagem. O descanso adequado melhora a memória, a concentração e o raciocínio lógico.

Recomendações para a Direção da Escola
Incentive o estudo diário estruturado – Crie programas de acompanhamento (como tutoriais ou salas de estudo monitoradas) para que os alunos estudem pelo menos 2 horas por dia fora da sala de aula.

Monitore e valorize a frequência – Implemente um sistema de alerta para alunos com baixa presença e premie aqueles com frequência acima de 95%. A presença precisa ser tratada como prioridade.

Eduque sobre a importância do sono – Inclua palestras ou materiais sobre higiene do sono, mostrando que dormir menos de 5 horas prejudica tanto quanto não estudar. Evite atividades escolares muito cedo ou muito tarde que prejudiquem o descanso.

Resumo executivo:

Estudar mais, ir às aulas com regularidade e dormir bem são os três hábitos que mais fazem a diferença nas notas. Focar nesses três pilares pode elevar significativamente o desempenho médio da escola.
