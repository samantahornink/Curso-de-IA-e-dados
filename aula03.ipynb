`````python
#Apelido
import pandas as pd
import numpy as np
`````

`````python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]


array_produtos = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
print(faturamento)
`````

`````python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]
nome_produtos = ["Pastel", "Caçulinha", "Refrigerante","Combos"]


array_produtos = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
for nome, preco in zip(nome_produtos,faturamento):
    print(f"{nome} : {preco}")
`````

### Exercício de fixação
- Criar umas lista que irá possuir 10 nomes de produtos.
- Criar uma lista que irá possuir o preço de custo (atacado).
- Criar uma lista que irá possuir o preço de venda (varejo).
- Ao final, exibir o lucro obtido em cada produto.

`````python
Nome_produtos = ["Base líquida", "Corretivo", "Batom líquido", "Lápis de olho", "Hidratante corporal", "Protetor solar", "Perfume Kaiak","Máscara de cílios", "Creme de barbear", "Hidratante para lábios" ]
Preco_custo = [42,35,30,25,50,55,120,42,40,32]
Preco_venda = [82,68,60,50,90,100,200,84,80,64]

array_preco_custo = np.array(Preco_custo)
array_preco_venda = np.array(Preco_venda)

Lucro = array_preco_venda - array_preco_custo
for name, preco in zip(Nome_produtos,Lucro):
    print(f"Lucro de produtos vendidos: {name}:{preco}")
    




    
`````
