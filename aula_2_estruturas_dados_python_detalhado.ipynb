## 1. O que são estruturas de dados?

Estruturas de dados são formas de organizar informações dentro do programa.

Em vez de guardar tudo em variáveis soltas, podemos agrupar dados em coleções.

### Exemplo
Em vez de fazer:
```python
temperatura1 = 70
temperatura2 = 75
temperatura3 = 80
```

Podemos fazer:
```python
temperaturas = [70, 75, 80]
```

### Por que isso é importante?
Porque facilita:
- organização
- leitura do código
- repetição
- análise de dados
- criação de algoritmos


## 2. Listas

A **lista** é uma das estruturas mais usadas em Python.

Ela serve para guardar vários valores em uma única variável.

### Características da lista
- usa colchetes `[ ]`
- pode guardar vários elementos
- mantém a ordem
- pode ser alterada


````python
# Criando uma lista
temperaturas = [68, 72, 75, 81, 90]

# Mostrando a lista inteira
print(temperaturas)
````

### Explicação
- `temperaturas` é o nome da lista
- os valores ficam entre colchetes
- cada valor é separado por vírgula


## 3. Acessando elementos da lista

Cada posição da lista tem um índice.

### Atenção
Em Python, a contagem começa no zero.


````python
temperaturas = [68, 72, 75, 81, 90]

print(temperaturas[0])  # primeiro valor
print(temperaturas[1])  # segundo valor
print(temperaturas[4])  # quinto valor
````

### Explicação
- `temperaturas[0]` acessa o primeiro elemento
- `temperaturas[1]` acessa o segundo
- o índice indica a posição do elemento


## 4. Alterando valores da lista

Uma lista pode ser modificada.


````python
temperaturas = [68, 72, 75, 81, 90]

# Alterando o valor da posição 2
temperaturas[2] = 76

print(temperaturas)
````

### Conceito importante
Listas são **mutáveis**.
Isso significa que seus elementos podem ser alterados depois da criação.


## 5. Adicionando elementos em listas

Podemos acrescentar novos dados usando `append()`.


````python
temperaturas = [68, 72, 75]

temperaturas.append(80)
temperaturas.append(85)

print(temperaturas)
````

### Explicação
- `append()` adiciona um novo elemento ao final da lista
- isso é muito útil quando os dados vão sendo recebidos ao longo do programa


## 6. Removendo elementos da lista

Também é possível remover valores.


````python
temperaturas = [68, 72, 75, 80]

temperaturas.remove(72)

print(temperaturas)
````

### Explicação
- `remove(72)` procura o valor 72 e remove esse elemento da lista
- a lista fica menor após a remoção


## 7. Percorrendo listas com `for`

Uma das maiores vantagens das listas é que podemos percorrê-las facilmente.


````python
temperaturas = [68, 72, 75, 81, 90]

for temperatura in temperaturas:
    print(f"Temperatura registrada: {temperatura}°C")
````

### Explicação
- o `for` percorre a lista elemento por elemento
- a variável `temperatura` recebe um valor por vez
- isso permite analisar vários dados com poucas linhas de código


## 8. Exemplo de lista com contexto industrial

Vamos imaginar uma lista com produções diárias.


````python
producoes = [120, 95, 140, 110, 150]

for producao in producoes:
    if producao >= 120:
        print(f"Produção {producao}: alta")
    else:
        print(f"Produção {producao}: abaixo do esperado")
````

## 9. Tuplas

A **tupla** é parecida com a lista, mas possui uma diferença importante:
ela **não pode ser alterada** depois de criada.

### Características da tupla
- usa parênteses `( )`
- mantém a ordem
- não é mutável


````python
# Criando uma tupla
cores_status = ("verde", "amarelo", "vermelho")

print(cores_status)
````

### Quando usar tuplas?
Quando queremos guardar dados que não devem ser alterados facilmente.

Exemplos:
- dias da semana
- meses do ano
- categorias fixas
- níveis de status


````python
cores_status = ("verde", "amarelo", "vermelho")

print(cores_status[0])
print(cores_status[1])
````

### Observação
Podemos acessar elementos da tupla pelo índice, assim como fazemos com listas.


## 10. Diferença entre lista e tupla

### Lista
- mutável
- ideal quando os dados mudam

### Tupla
- imutável
- ideal quando os dados devem ficar fixos


````python
# Lista: pode mudar
lista_exemplo = [1, 2, 3]
lista_exemplo[0] = 10
print(lista_exemplo)

# Tupla: não pode mudar
tupla_exemplo = (1, 2, 3)
print(tupla_exemplo)
````

## 11. Dicionários

O **dicionário** guarda dados no formato:
- chave
- valor

Ele é muito útil quando queremos relacionar um identificador a uma informação.


````python
# Criando um dicionário
maquina = {
    "nome": "Prensa 01",
    "temperatura": 78,
    "status": "OK"
}

print(maquina)
````

### Características do dicionário
- usa chaves `{ }`
- cada item possui uma chave e um valor
- ajuda muito a organizar dados com significado


## 12. Acessando valores do dicionário

Para acessar um valor, usamos a chave.


````python
maquina = {
    "nome": "Prensa 01",
    "temperatura": 78,
    "status": "OK"
}

print(maquina["nome"])
print(maquina["temperatura"])
print(maquina["status"])
````

### Explicação
- `"nome"` é a chave
- `"Prensa 01"` é o valor associado
- o dicionário facilita guardar dados com contexto


## 13. Alterando valores do dicionário

Assim como a lista, o dicionário também pode ser alterado.


````python
maquina = {
    "nome": "Prensa 01",
    "temperatura": 78,
    "status": "OK"
}

maquina["temperatura"] = 85
maquina["status"] = "ATENCAO"

print(maquina)
````

## 14. Adicionando novos pares chave-valor

Também podemos acrescentar novas informações.


````python
maquina = {
    "nome": "Prensa 01",
    "temperatura": 78
}

maquina["status"] = "OK"
maquina["setor"] = "Estamparia"

print(maquina)
````

## 15. Percorrendo dicionários

Podemos percorrer um dicionário para ver suas chaves e valores.


````python
maquina = {
    "nome": "Prensa 01",
    "temperatura": 78,
    "status": "OK"
}

for chave, valor in maquina.items():
    print(f"{chave} -> {valor}")
````

### Explicação
- `items()` retorna pares de chave e valor
- isso é muito útil quando queremos mostrar ou analisar todas as informações do dicionário


## 16. Conjuntos

O **conjunto** (`set`) é uma coleção que:
- não mantém ordem garantida
- não permite valores repetidos

Ele é útil quando queremos eliminar duplicatas.


````python
# Criando um conjunto
sensores = {"S1", "S2", "S3", "S2", "S1"}

print(sensores)
````

### O que aconteceu?
Mesmo que `S1` e `S2` apareçam repetidos, o conjunto guarda apenas valores únicos.


## 17. Criando conjunto a partir de lista

Uma forma muito prática de remover duplicatas é transformar uma lista em conjunto.


````python
leituras = [70, 72, 72, 75, 80, 80, 81]

leituras_unicas = set(leituras)

print(leituras_unicas)
````

### Aplicação
Esse recurso pode ser útil quando queremos:
- identificar valores únicos
- remover repetições
- preparar dados


## 18. Comparando estruturas

### Lista
- ordenada
- mutável
- aceita repetição

### Tupla
- ordenada
- imutável
- aceita repetição

### Dicionário
- chave e valor
- mutável
- muito útil para dados organizados por nome

### Conjunto
- não permite repetição
- útil para valores únicos


## 19. Quando usar cada uma?

### Use lista quando:
- houver vários valores do mesmo tipo
- os dados puderem mudar
- a ordem for importante

### Use tupla quando:
- os dados forem fixos

### Use dicionário quando:
- cada informação tiver um nome
- for importante relacionar chave e valor

### Use conjunto quando:
- o foco for remover duplicatas ou obter valores únicos

## 21. Introdução básica a funções

Funções servem para agrupar comandos que realizam uma tarefa.

### Vantagens
- organizam o código
- evitam repetição
- facilitam reaproveitamento


````python
# Criando uma função simples
def saudacao():
    print("Olá! Esta é uma função em Python.")

# Chamando a função
saudacao()
````

### Explicação linha a linha
- `def` é a palavra usada para criar uma função
- `saudacao` é o nome da função
- os parênteses podem receber parâmetros
- o bloco recuado é o corpo da função
- para usar a função, precisamos chamá-la pelo nome


## 22. Função com parâmetro

Uma função pode receber informações de entrada.


````python
def mostrar_temperatura(valor):
    print(f"Temperatura recebida: {valor}°C")

mostrar_temperatura(78)
mostrar_temperatura(91)
````

### Conceito
- `valor` é o parâmetro da função
- quando chamamos a função, enviamos um dado para ela trabalhar


## 23. Função que percorre uma coleção

Agora vamos juntar dois temas da aula:
- coleções
- funções


````python
def analisar_temperaturas(lista_temperaturas):
    for temperatura in lista_temperaturas:
        if temperatura >= 90:
            print(f"{temperatura}°C -> crítico")
        elif temperatura >= 75:
            print(f"{temperatura}°C -> atenção")
        else:
            print(f"{temperatura}°C -> normal")

temperaturas = [68, 74, 79, 91, 85]
analisar_temperaturas(temperaturas)
````

### O que essa função faz?
- recebe uma lista
- percorre a lista com `for`
- classifica cada valor
- mostra o resultado

### Relação com o curso
Esse tipo de função já se aproxima da lógica usada em análise inicial de dados.
